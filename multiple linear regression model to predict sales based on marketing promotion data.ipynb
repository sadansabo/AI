````python
import pandas as pd

# Assuming your original dataset is called 'data'
data = data.dropna()

# Encode categorical variables properly
data = pd.get_dummies(data, columns=['TV', 'Influencer'], drop_first=True)

# Check the actual column names
print(data.columns)

````

````python
import statsmodels.formula.api as smf

# Adjusted formula using actual dummy column names
formula = 'Sales ~ Radio + Q("Social Media") + TV_Low + TV_Medium + Influencer_Mega + Influencer_Micro + Influencer_Nano'

model = smf.ols(formula=formula, data=data).fit()
print(model.summary())

````

````python
import seaborn as sns
import statsmodels.api as sm

# Predicted values and residuals
data['fitted'] = model.fittedvalues
data['residuals'] = model.resid

# Linearity check: fitted vs actual
sns.scatterplot(x=data['fitted'], y=data['Sales'])
plt.xlabel('Fitted Values')
plt.ylabel('Actual Sales')
plt.title('Linearity Check')
plt.show()

# Q-Q plot for normality
sm.qqplot(data['residuals'], line='s')
plt.title('Q-Q Plot of Residuals')
plt.show()
````
